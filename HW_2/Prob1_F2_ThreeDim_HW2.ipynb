```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f2(x, y):
    return 1 - (x ** 2 + y ** 2)

x = np.linspace(-50, 50, 50) 
y = np.linspace(-50, 50, 40)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)

p2Z = Z.max()

res = np.where(Z == p2Z)
coords = list(zip(res[0], res[1]))

row = coords[0][0]
col = coords[0][1]

p2X = X[row][col]
p2Y = Y[row][col]

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.plot([p2X], p2Y, p2Z,'ro') 
ax.text(p2X, p2Y, p2Z, "p2", fontsize=12)

```
