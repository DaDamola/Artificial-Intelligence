```
%tensorflow_version 2.x
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

counter = [0] * 10
res0, res1, res2, res3, = np.zeros((28, 28)), np.zeros((28, 28)),  np.zeros((28, 28)), np.zeros((28, 28))
res4, res5, res6, res7, = np.zeros((28, 28)), np.zeros((28, 28)),  np.zeros((28, 28)), np.zeros((28, 28))
res8, res9 = np.zeros((28, 28)), np.zeros((28, 28))

#calculate digit counts and sums 
for i in range(len(train_images)):
  if train_labels[i] == 0:
    res0 += train_images[i]
    counter[0] += 1 
  elif train_labels[i] == 1:
    res1 += train_images[i]
    counter[1] += 1 
  elif train_labels[i] == 2:
    res2 += train_images[i]
    counter[2] += 1 
  elif train_labels[i] == 3:
    res3 += train_images[i]
    counter[3] += 1 
  elif train_labels[i] == 4:
    res4 += train_images[i]
    counter[4] += 1 
  elif train_labels[i] == 5:
    res5 += train_images[i]
    counter[5] += 1 
  elif train_labels[i] == 6:
    res6 += train_images[i]
    counter[6] += 1 
  elif train_labels[i] == 7:
    res7 += train_images[i]
    counter[7] += 1 
  elif train_labels[i] == 8:
    res8 += train_images[i]
    counter[8] += 1 
  elif train_labels[i] == 9:
    res9 += train_images[i]
    counter[9] += 1 

#display digits
plt.figure(figsize=(4, 4))
plt.imshow(res0 / counter[0])
plt.show()

plt.figure(figsize=(4, 4))
plt.imshow(res1 / counter[1])
plt.show()

plt.figure(figsize=(4, 4))
plt.imshow(res2 / counter[2])
plt.show()

plt.figure(figsize=(4, 4))
plt.imshow(res3 / counter[3])
plt.show()

plt.figure(figsize=(4, 4))
plt.imshow(res4 / counter[4])
plt.show()

plt.figure(figsize=(4, 4))
plt.imshow(res5 / counter[5])
plt.show()

plt.figure(figsize=(4, 4))
plt.imshow(res6 / counter[6])
plt.show()

plt.figure(figsize=(4, 4))
plt.imshow(res7 / counter[7])
plt.show()

plt.figure(figsize=(4, 4))
plt.imshow(res8 / counter[8])
plt.show()

plt.figure(figsize=(4, 4))
plt.imshow(res9 / counter[9])
plt.show()


```
